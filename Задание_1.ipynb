*Найдите и перечислите все логические противоречия и недочеты в предоставленном ТЗ. Объясните, в чем заключается проблема для каждого пункта.*

1. Противоречие между пунктами 2 и 9. В пункте 2 указано, что пользователь может изменить количество товара не менее чем до 1, а удаление товара осуществляется отдельной кнопкой. В пункте 9 сказано, что при уменьшении количества товара до 0 он удаляется из корзины. Нужна однозначность в формулировке: либо уменьшение до 0 запрещено, либо удаление через 0 допустимо.  
2. Противоречие между пунктами 1, 3, 4. В пункте 1 указано, что пользователь может добавить до 10 единиц одного товара. В пункте 3 сказано, что в корзине может находиться не более 5 различных товаров. Следовательно, максимальное количество товаров в корзине должно быть 50. Но в пункте 4, говорится про ограничение в 20 единиц.    
3. Избыточность пункта 5. В пункте 5 сказано, что товары в корзине могут быть разные. Но эта информация вытекает из пункта 3.   
4. Противоречие между пунктами 7 и 13. В пункте 7 указано, что цена товара фиксируется на момент добавления в корзину и не меняется.
В пункте 13 сказано, что при изменении цены в каталоге система должна автоматически обновить цену в корзине у всех пользователей. Эти требования взаимоисключают друг друга.  
5. Неоднозначность логики удаления товара пункты 2 и 9. Необходимо однозначно определить логику удаления товара из корзины. Посредством уменьшения количества товара до 0 или существованием отдельной кнопки удаления товара.  
6. Недостаточная детализация сообщения о превышении лимита пункт 6. Неясно, в каком именно случае должно отображаться данное сообщение (лимит на количество одного товара, лимит на общее количество товаров). Скорее всего имеется в виду все перечисленные случаи, но для пользователя было бы удобнее понимать о каком превышении идёт речь.  
7. Недостаточная детализация требований к рекламе в корзине пункт 11. Сказано, что реклама должна отображаться каждый будний день по утрам и вечерам, но не уточнены: временные интервалы, часовой пояс, зависит ли показ от региона пользователя.  
8. Нарушение нумерации. Всего должно быть 12 пунктов. Но в ТЗ 12 пункт пропущен.  


*Предложите конкретные исправления для устранения этих противоречий - напишите свою версию этого фрагмента ТЗ, которая должна быть логически завершенной и непротиворечивой.*

1. Пользователь может добавить в корзину от 1 до 10 единиц одного товара.  
2. В корзине может находиться не более 5 различных товаров.  
3. Суммарное количество всех товаров в корзине не может превышать 50 единиц.  
4. Пользователь может изменять количество каждого товара в корзине в диапазоне от 1 до максимально допустимого значения.  
5. Для удаления товара из корзины используется отдельная кнопка «Удалить». Уменьшение количества товара до 0 не допускается.  
6. При попытке добавить товар или изменить его количество с превышением установленных ограничений система отображает сообщение: «Лимит корзины превышен».  
7. Если цена на товар изменилась в каталоге, система должна автоматически обновить ее в корзине у всех пользователей.   
8. На странице корзины отображается: список добавленных товаров, количество каждого товара, цена за единицу товара, общая стоимость позиции, общая стоимость всех товаров в корзине.  
9. В корзине может отображаться реклама других товаров.  
10. Реклама в корзине отображается по будним дням в следующие временные интервалы: утро(c 09:00 до 12:00), вечер(c 18:00 до 21:00 ) 
Время определяется по часовому поясу пользователя.
